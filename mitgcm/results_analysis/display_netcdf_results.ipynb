```python
import xarray as xr
import glob
import netCDF4 as nc
import numpy as np
import os

import matplotlib.pyplot as plt
import holoviews as hv
hv.extension('bokeh')  # You can also use 'matplotlib' or 'plotly'

#%matplotlib notebook
# seaborn for interactive plots or bokee or hvplot
```

```python
path_results_folder = r"D:\geneva_200m\results_netcdf"
merged_results_file = os.path.join(path_results_folder, '3Dsnaps.0000000000.t001.nc')

var_to_plot = 'THETA'
var_u = 'UVEL'
var_v = 'VVEL'
var_depth = 'Zmd000100'
```

xr_datasets = []

```python
ds_to_plot = xr.open_dataset(merged_results_file)
```

```python
ds_to_plot['T'][-1]
```

ds_to_plot[var_to_plot].values.max()

```python
i_t = 0
T_plot = ds_to_plot['T'].values[i_t]
Z_plot = ds_to_plot[var_depth].values[0]
```

```python
subsetting_factor = 20
X_trimmed = ds_to_plot['X'][::subsetting_factor]
Y_trimmed = ds_to_plot['Y'][::subsetting_factor]
U_trimmed = ds_to_plot[var_u].sel(T=T_plot, Zmd000100=Z_plot)[:,1:][::subsetting_factor,::subsetting_factor]
V_trimmed = ds_to_plot[var_v].sel(T=T_plot, Zmd000100=Z_plot)[1:,:][::subsetting_factor,::subsetting_factor]
```

```python
ds_to_plot['T'].values[-1]
```

```python
ds_to_plot[var_to_plot].sel(T=ds_to_plot['T'].values[-1], Zmd000100=Z_plot)['T']
```

```python
i_t = 1
T_plot = ds_to_plot['T'].values[i_t]
xr_plot = ds_to_plot[var_to_plot].sel(T=T_plot, Zmd000100=Z_plot)
xr_plot = xr_plot.where(xr_plot != 0, np.nan)
xr_plot.plot(cmap='jet', figsize=(15,5), vmin=24, vmax=27)
plt.axis('equal')
plt.quiver(X_trimmed, Y_trimmed, U_trimmed, V_trimmed, scale=3)
```

```python
# Convert to magnitude and angle
mag = np.sqrt(U_trimmed.values**2 + V_trimmed.values**2)
angle = (np.pi/2.) - np.arctan2(U_trimmed.values/mag, V_trimmed.values/mag)

vector_field = hv.VectorField((X_trimmed, Y_trimmed, angle, mag)).opts(
    color='green',  # Arrow color
    magnitude='Magnitude',  # Set arrow length to represent speed
    arrow_heads=True,  # Display arrow heads
    scale=0.5,  # Scale of arrows
    width=600, height=400,
    title='Water Velocity Field'
)
```

```python
image = hv.Image(xr_plot, kdims=[xr_plot.dims[1], xr_plot.dims[0]])
image.opts(tools=['hover'], colorbar=True, cmap='viridis', width=1600, height=800, title=xr_plot.attrs['description']+' '+ str(T_plot))
image * vector_field
```
