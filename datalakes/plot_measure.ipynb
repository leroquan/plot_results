```python
import xarray as xr
```

```python
measure_paths = ['../data/datalakes/geneva_idronaut/2023.nc', 
                 '../data/datalakes/geneva_idronaut/2024.nc']
```

```python
xr_measures = []
for path in measure_paths:
    xr_measures.append(xr.open_dataset(path))
```

```python
for ds in xr_measures:
    ds['temp'].sel(depth=-10, method='nearest').plot()
```

```python
for ds in xr_measures:
    ds['temp'].sel(depth=-10, method='nearest').resample(time='1D').mean().plot()
```

```python
xr_measures[1]['temp'].sel(depth=-10, method='nearest').resample(time='1D').mean().plot()
```
